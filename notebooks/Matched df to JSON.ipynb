```python
df = pd.read_csv(r"C:\Users\Travis\OneDrive - College of Coastal Georgia\work\notebooks\coordinate matching notebooks\full_field_cluster.csv")
df.head()
```

```python
parameter_list = ['long','lat','plant_area','canopy_cover']

# geno_data = defaultdict(dict)
# treatment_data = defaultdict(dict)
# replicant_data = defaultdict(dict)
# ID_data = defaultdict(dict)
# date_data = defaultdict(dict)
# param_data = defaultdict(dict)

geno_list = df.genotype.unique().tolist()
if 'Green_Towers_BORDER' in geno_list:
    geno_list.remove('Green_Towers_BORDER')
cnt = 0
geno_data = {}
for geno in geno_list:
    df2 = df[df['genotype'] == geno]
    treatment_list = df2.treatment.unique().tolist()
    
    treatment_data = {}
    for treatment in treatment_list:
        df3 = df2[df2['treatment'] == treatment]
        replicant_list = df3['plot'].unique().tolist()
        
        replicant_data = {}
        for replicant in replicant_list:
            df4 = df3[df3['plot'] == replicant]
            ID_list = df4.plant_name.unique().tolist()

            ID_data = {}
            for ID in ID_list:
                df5 = df4[df4['plant_name'] == ID]
                date_list = df5.date.unique().tolist()
                
#                 print(list(ID.split(','))[-1].replace(')',''))
                date_data = {}
                for date in date_list:
                    df6 = df5[df5['date'] == date]
                    param_data = {}
                    
                    for parameter in parameter_list:
                        
#                         try:
#                             geno_data[geno][treatment][replicant][ID][date][parameter] == df6['plant_area'].iloc[0]
                        
#                         except:
#                             geno_data[geno][treatment][replicant][ID][date][parameter] == 'NA'
                        
                        try:
                            param_data[f'{parameter}'] = df6[f'{parameter}'].iloc[0]
                        
                        except:
                            param_data[f'{parameter}'] = 'NA'
                        
                    date_data[f'{date}'] = param_data
                ID = str(ID).replace(',','_').replace(' ','').replace("'","").replace('(','').replace(')','')
                ID_data[f'{ID}'] = date_data

            replicant_data[f'{replicant}'] = ID_data
        
        treatment_data[f'{treatment}'] = replicant_data
    
    geno_data[f'{geno}'] = treatment_data
                
with open('full_field_cluster_all_params_new_name.json', 'w') as outfile:
    
    json.dump(geno_data, outfile)                        
```
